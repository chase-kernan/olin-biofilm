```python
from numpy import *
from matplotlib.pyplot import *
from matplotlib import pylab
from matplotlib import rc
from matplotlib.ticker import ScalarFormatter
from StringIO import StringIO

%matplotlib inline
rc('font', family='Arial', size='10')
rc('text', usetex=False)
rc('savefig', bbox='tight', pad_inches=0.03)
```

```python
data_str = """Time	Aerobic A	Aerobic B	Aerobic C	Anaerobic A	Anaerobic B	Anaerobic C
0.1	0.034	0.026	0.027	0.032	0.026	0.034
8.25	0.064	0.044	0.046	0.059	0.047	0.062
23.5	0.163	0.127	0.129	0.128	0.111	0.142
31	0.227	0.189	0.188	0.19	0.169	0.208
46.5	0.336	0.292	0.301	0.399	0.361	0.454
51.5	0.375	0.33	0.331	0.51	0.45	0.543
72.5	0.473	0.417	0.444	1.086	1.077	1.222
77.5	0.528	0.435	0.566	1.281	1.271	1.33
146.5	0.675	0.625	0.846	1.316	1.291	1.288"""
data = recfromcsv(StringIO(data_str), delimiter='\t')

from string import capwords
fields = [(f, capwords(f.replace('_', ' ').replace('bic', '.')))
          for f in sorted(data.dtype.fields.keys()) 
          if f != 'time']
```

```python
aerobic = vstack((data['aerobic_a'], data['aerobic_b'], data['aerobic_c']))
anaerobic = vstack((data['anaerobic_a'], data['anaerobic_b'], data['anaerobic_c']))
```

```python
from scipy.stats import linregress

time = tile(data['time'], 3)
flat_aerobic = hstack((aerobic[0, :], aerobic[1, :], aerobic[2, :]))
flat_anaerobic = hstack((anaerobic[0, :], anaerobic[1, :], anaerobic[2, :]))

print time
print flat_aerobic
print flat_anaerobic

fit_filter = time < 70

def do_exp_fit(series):
    fit_time = time[fit_filter]
    values = series[fit_filter]
    regress = linregress(fit_time, log(values))
    
    print 'y = {}e^({}t)'.format(exp(regress[1]), regress[0])
    print 'R^2', regress[2]**2
    
    return fit_time, exp(regress[0]*fit_time + regress[1])

print "Aerobic"
fit_time, fit_aerobic = do_exp_fit(flat_aerobic)
print "Anaerobic"
fit_time, fit_anaerobic = do_exp_fit(flat_anaerobic)
```

Combined, Logarithmic
=====================

```python
figure(figsize=(3.27, 5.0))
hold(True)

styles = 'ro', 'ro', 'ro', 'bx', 'bx', 'bx'

for (field, label), style in zip(fields, styles):
    semilogy(data['time'], data[field], style, label=label,
             mfc='none', mew=1, mec=style[0])

semilogy(fit_time, fit_aerobic, 'r-', label='Aerobic Fit')
semilogy(fit_time, fit_anaerobic, 'b-', label='Anaerobic Fit')

xlim([-5, 150])
ylim([0.02, 1.5])
yticks([0.03, 0.1, 0.3, 1])
gca().yaxis.set_major_formatter(ScalarFormatter())

# Combine A and B labels
handles, labels = gca().get_legend_handles_labels()
handles = [handles[i] for i in [0, 6, 3, 7]]
labels = ['Aerobic', 'Aerobic Fit', 'Anaerobic', 'Anaerobic Fit']
legend(handles, labels, loc='lower right', numpoints=1)

xlabel('Time [hr]')
ylabel('Optical Density at 660 nm')

savefig('plots/images/growth_rate/combined_log.tiff', dpi=400)
```
