# Importing

```python
import pandas as pd
import tables
```

```python
files = ['results/{}_sample.h5'.format(x) for x in ['d', 'h', 'j', 'k', 'm', 'n']]
```

```python
specs = pd.read_hdf(files[0], 'specs')
for f in files[1:]:
    specs = specs.append(pd.read_hdf(f, 'specs'), ignore_index=True)
```

```python
spec_uuid_to_id = {}
for id, row in specs.iterrows():
    spec_uuid_to_id[row['uuid']] = id
```

```python
results = None
for f in files:
    print(f)
    print('---------------------')
    
    table = tables.open_file(files[0], 'r+')
    images = table.get_node('/_results_image_32_256')
    mapping = table.get_node('/results')
    
    result_uuid_to_spec_id = {}
    for row in mapping.iterrows():
        result_uuid_to_spec_id[row['uuid']] = spec_uuid_to_id[row['spec_uuid']]
    print('mapped')
    
    
    spec_ids = np.empty(mapping.nrows, int)
    biomasses = np.empty(mapping.nrows, float)
    coverages = []
    for i, row in enumerate(images.iterrows()):
        spec_ids[i] = result_uuid_to_spec_id[row['uuid']]
        biomasses[i] = row['data'].sum()/256.0
        
        unnormalized = row['data'].sum(axis=1)/256.0
        coverages.append(unnormalized/unnormalized.sum())
    
    file_results = pd.DataFrame({'spec_id': spec_ids, 'biomass': biomasses, 'coverage': coverages})
    if results is not None:
        results = results.append(file_results, ignore_index=True)
    else:
        results = file_results
        
```
