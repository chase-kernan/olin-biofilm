```python
import numpy as np
from numpy import *
from scipy import interpolate
import os
import csv
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from matplotlib.pyplot import *
from matplotlib.cm import Blues_r, Blues, gist_heat, hot, bone
from matplotlib import pylab
set_cmap('hot')
pylab.rcParams['figure.figsize'] = 10, 10
%matplotlib inline
```

Parameter Ranges
----------------

```python
param_ranges = {
    'stop_on_mass': 4000,
    'stop_on_time': 60000,
    'boundary_layer': 7,
    'light_penetration': (0.1, 50),
    'tension_power': (0.1, 3),
    'distance_power': (0.1, 3),
    'initial_cell_spacing': 2,
    'media_ratio': (0.01, 0.60),
    'media_monod': (0.01, 1),
    'light_monod': (0.01, 1)
}

param_features = []
for param, r in param_ranges.iteritems():
    if isinstance(r, tuple):
        param_features.append(param)
```

Load Results
------------

```python
features = []
values = []
for name in os.listdir('dump'):
    if not name.endswith('.h5.dump'): continue
    with open('dump/'+name, 'r') as csv_file:
        for row in csv.DictReader(csv_file):
            features.append([float(row[param]) for param in param_features])
            values.append(float(row['hsa']))

features = np.array(features, dtype=float)
values = np.array(values, dtype=float)
```

```python
def regress_scikit(features, values):
    reg = LinearRegression()
    reg.fit(features, values, n_jobs=-1)
    print 'Coefficients'
    print '------------'
    for param, coeff in zip(param_features, reg.coef_):
        print '  ', param + ':', coeff
    print '------------'
    print 'Intercept:', reg.intercept_
    print 'R^2:', reg.score(features, values)
    
def regress_sm(features, values):
    features = sm.add_constant(features, prepend=False)
    model = sm.OLS(values, features)
    model.exog_names[:-1] = param_features
    result = model.fit()
    print result.summary()
```

Linear Regression
=================

```python
regress_sm(features, values)
```

High light penetration regression
---------------------------------
p > 6

```python
high_light = features[:,-1] > 20
regress_sm(features[high_light, :], values[high_light])
```

High light penetration and uptake ratio regression
---------------------------------
p > 6 and r > 0.8

```python
high_light_and_media = (features[:,-1] > 6) & (features[:, 1] > 0.4)
regress_sm(features[high_light_and_media, :], values[high_light_and_media])
```

```python
def plot_phase(x, y, z, num_cells=10, **plot_args):
    x_min, x_max = x.min(), x.max()
    y_min, y_max = y.min(), y.max()
    
    try:
        num_x, num_y = num_cells
    except TypeError:
        num_x, num_y = num_cells, num_cells
    
    x_grid, y_grid = np.mgrid[x_min:x_max:num_x*1j,
                              y_min:y_max:num_y*1j]
            
    z_sum = np.histogram2d(x, y, num_cells, weights=z)[0]
    z_count = np.histogram2d(x, y, num_cells)[0]
    z_count[z_count == 0] = 1
    z_grid = z_sum/z_count
    
    pcolormesh(x_grid, y_grid, z_grid, **plot_args)
    xlim(x_min, x_max)
    ylim(y_min, y_max)
```

Parameter Comparisons
=====================

Absolute H value
----------------

```python
n = len(param_features)
#v_min, v_max = values.min(), values.max()
v_min = 0
v_max = 3 # averaging brings most values down

for i, param1 in enumerate(param_features):
    for j, param2 in enumerate(param_features):
        subplot(n, n, i*n + j + 1)
        plot_phase(features[:, j], features[:, i], values,
                   vmin=v_min, vmax=v_max, cmap=Blues)
        
        tick_params(which='both', bottom='off', top='off', left='off', 
                    right='off', labelbottom='off', labelleft='off')
        if j == 0: ylabel(param1)
        if i == n-1: xlabel(param2)
gcf().set_size_inches(10,10)
savefig('plots/images/param_space_absoulte.png', dpi=300)
show()
```
