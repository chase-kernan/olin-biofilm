```python
import numpy as np
import os
import csv
from sklearn.linear_model import LinearRegression
```

```python
param_ranges = {
    'stop_on_mass': 4000,
    'stop_on_time': 20000,
    'boundary_layer': 5,
    'light_penetration': (0.1, 32),
    'tension_power': (0.1, 4),
    'distance_power': (0.1, 4),
    'initial_cell_spacing': 2,
    'media_ratio': (0.01, 1.2),
    'media_monod': (0.01, 1),
    'light_monod': (0.01, 1)
}

param_features = []
for param, r in param_ranges.iteritems():
    if isinstance(r, tuple):
        param_features.append(param)
```

```python
features = []
values = []
for name in os.listdir('dump'):
    with open('dump/'+name, 'r') as csv_file:
        for row in csv.DictReader(csv_file):
            #row['media_ratio'] = 1/(25*float(row['media_ratio']))
            features.append([float(row[param]) for param in param_features])
            values.append(float(row['hsa']))

print len(features)
features = np.array(features, dtype=float)
print features.shape
values = np.array(values, dtype=float)
```

```python
regression = LinearRegression()
regression.fit(features, values, n_jobs=-1)
print 'Features', param_features
print 'Coeff', regression.coef_
print 'Intercept', regression.intercept_
print 'Params', regression.get_params()
print 'R2', regression.score(features, values)
```

```python
regression = LinearRegression()
high_light = features[:,-1] > 6
regression.fit(features[high_light, :], values[high_light], n_jobs=-1)
print 'Features', param_features
print 'Coeff', regression.coef_
print 'Intercept', regression.intercept_
print 'Params', regression.get_params()
print 'R2', regression.score(features, values)
```

```python
regression = LinearRegression()
high_light_and_media = (features[:,-1] > 6) & (features[:, 1] > 0.8)
regression.fit(features[high_light_and_media, :], values[high_light_and_media], n_jobs=-1)
print 'Features', param_features
print 'Coeff', regression.coef_
print 'Intercept', regression.intercept_
print 'Params', regression.get_params()
print 'R2', regression.score(features, values)
```

```python
from scipy import interpolate

def plot_phase(x, y, z, num_cells=10, **plot_args):
    x_min, x_max = x.min(), x.max()
    y_min, y_max = y.min(), y.max()
    
    try:
        num_x, num_y = num_cells
    except TypeError:
        num_x, num_y = num_cells, num_cells
        
    grid = np.mgrid[x_min:x_max:num_x*1j,
                    y_min:y_max:num_y*1j]
    #interp = interpolate.griddata(np.hstack((x.reshape((x.size, 1)), 
    #                                         y.reshape((y.size, 1)))), 
    #                              z,
    #                              np.vstack((grid[0].flat, grid[1].flat)),
    #                              'linear')
    #z_grid = np.reshape(interp, grid[0].shape)
    z_sum = np.histogram2d(x, y, num_cells, weights=z)[0]
    z_count = np.histogram2d(x, y, num_cells)[0]
    z_count[z_count == 0] = 1
    z_grid = z_sum/z_count
    
    pcolormesh(grid[0], grid[1], z_grid, **plot_args)
    xlim(x_min, x_max)
    ylim(y_min, y_max)
```

```python
n = len(param_features)
v_min, v_max = values.min(), values.max()
span = v_max - v_min
v_min += span*0
v_max -= span*0.6

for i, param1 in enumerate(param_features):
    for j, param2 in enumerate(param_features):
        subplot(n, n, i*n + j + 1)
        plot_phase(features[:, j], features[:, i], values,
                   vmin=v_min, vmax=v_max, cmap=Blues)
        tick_params(which='both', bottom='off', top='off', left='off', 
                    right='off', labelbottom='off', labelleft='off')
        if j == 0:
            ylabel(param1)
        if i == n-1:
            xlabel(param2)
gcf().set_size_inches(10,10)
savefig('plots/images/param_space_absoulte.png', dpi=300)
show()
```
